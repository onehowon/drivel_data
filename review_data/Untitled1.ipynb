```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

# Chrome 드라이버 생성
driver = webdriver.Chrome()

try:
    # 웹페이지 접속
    url = "https://www.carisyou.com/car/"
    driver.get(url)
    time.sleep(3)  # 페이지가 로드될 때까지 잠시 대기

    # 데이터를 담을 집합 생성 (중복 방지를 위해 집합 사용)
    data = set()
    count = 0

    while True:
        # HTML 파싱
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # 차량 정보 추출
        cars = soup.find_all('a', class_='hover_line')
        for car in cars:
            text = car.text.strip()
            if text not in data:
                data.add(text)
                count += 1

            # 500개씩 데이터를 한 파일에 추가
            if count % 500 == 0:
                df = pd.DataFrame(list(data), columns=['Car Information'])
                df.to_csv('car_information.csv', index=False, mode='a', header=not count > 500)
                data.clear()  # 집합 초기화

        # '더 보기' 버튼이 없으면 반복 종료
        more_button = driver.find_elements(By.XPATH, '//a[@class="more_btn"]')
        if not more_button:
            break

        # '더 보기' 버튼 클릭
        more_button[0].click()
        time.sleep(2)  # 데이터가 로드될 때까지 잠시 대기

    # 마지막에 남은 데이터를 CSV 파일에 추가
    if data:
        df = pd.DataFrame(list(data), columns=['Car Information'])
        df.to_csv('car_information.csv', index=False, mode='a', header=not count > 500)

finally:
    # 작업 완료 후 브라우저 닫기
    driver.quit()
```
