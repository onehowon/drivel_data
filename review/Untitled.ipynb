```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# ChromeDriver 경로 지정
s = Service(r'C:\Users\sande\.cache\selenium\chromedriver\win64\126.0.6478.63\chromedriver.exe')
driver = webdriver.Chrome(service=s)

driver.get('https://map.naver.com/')

# 검색창 대기 및 입력
search = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, ".input_search"))
)
search.send_keys("대구 맛집")
search.send_keys(Keys.ENTER)

# 프레임 전환
WebDriverWait(driver, 10).until(
    EC.frame_to_be_available_and_switch_to_it((By.ID, 'searchIframe'))
)

# 가게 이름을 저장할 리스트
store_names = []

def get_store_names():
    # 원하는 요소(가게 이름) 찾기
    titles = WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".place_bluelink.TYaxT"))
    )

    for title in titles:
        store_names.append(title.text)

# 첫 페이지 가게 이름 수집
print("Page 1:")
get_store_names()

for page in range(2, 6):  # 2부터 5페이지까지
    try:
        # 페이지 번호 버튼 클릭
        next_page_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f"//a[@class='mBN2s ' and text()='{page}']"))
        )
        next_page_button.click()
        
        # 페이지가 완전히 로드될 때까지 대기
        time.sleep(2)  # 페이지 전환 대기
        
        # 프레임 다시 전환
        driver.switch_to.default_content()
        WebDriverWait(driver, 10).until(
            EC.frame_to_be_available_and_switch_to_it((By.ID, 'searchIframe'))
        )

        # 현재 페이지 가게 이름 수집
        print(f"Page {page}:")
        get_store_names()
        
    except Exception as e:
        print(f"Failed to go to page {page}: {e}")
        break

# 드라이버 종료
driver.quit()

# 수집한 데이터를 DataFrame으로 변환
df = pd.DataFrame(store_names, columns=["Store Name"])

# CSV 파일로 저장
df.to_csv("store_names.csv", index=False, encoding='utf-8-sig')

# XLSX 파일로 저장
df.to_excel("store_names.xlsx", index=False)

print("Data has been saved to 'store_names.csv' and 'store_names.xlsx'")
```
