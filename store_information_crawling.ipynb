```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException
from bs4 import BeautifulSoup
import time

# CSV 파일 읽기
df_stores = pd.read_csv('맛집_상호명.csv')  # 혹은 'euc-kr'

# 결과를 저장할 DataFrame 초기화
naver_res = pd.DataFrame(columns=['업체명', '업종', '주소', '전화번호', '썸네일'])

# 웹드라이버 초기화 및 기본 설정
driver = webdriver.Chrome()

def search_iframe():
    driver.switch_to.default_content()
    WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.ID, "searchIframe")))
    print("Switched to searchIframe")

def entry_iframe():
    driver.switch_to.default_content()
    try:
        WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.ID, "entryIframe")))
        for i in range(5):
            time.sleep(.5)
            try:
                driver.switch_to.frame(driver.find_element(By.ID, "entryIframe"))
                print("Switched to entryIframe")
                break
            except NoSuchElementException:
                pass
    except TimeoutException:
        print("TimeoutException: entryIframe not available.")

def extract_details(store_name):
    global naver_res
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    try:
        category = soup.select_one('span.lnJFt').text
    except:
        category = float('nan')

    try:
        address = soup.select_one('span.LDgIH').text
    except:
        address = float('nan')

    try:
        phone = soup.select_one('span.xlx7Q').text
    except:
        phone = float('nan')

    try:
        thumbnail = soup.select_one('img.K0PDV')['src']
    except:
        thumbnail = float('nan')

    # 데이터프레임에 추가
    new_row = pd.DataFrame({
        '업체명': [store_name], 
        '업종': [category], 
        '주소': [address], 
        '전화번호': [phone],
        '썸네일': [thumbnail]
    })
    naver_res = pd.concat([naver_res, new_row], ignore_index=True)
    print(f"Extracted details for {store_name}")

def crawling_main(store_name, city):
    global naver_res
    url = f"https://map.naver.com/p/search/{store_name} {city}"  # URL 구성
    driver.get(url)
    print(f"Navigating to {url}")

    try:
        # 세부 정보 페이지가 이미 로드되었는지 확인
        WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//*[@id="entryIframe"]')))
        entry_iframe()
        extract_details(store_name)  # 세부 정보 페이지에서 데이터 추출
    except TimeoutException:
        print(f"Details page not loaded directly for {store_name}, trying to click the store link")
        # 세부 정보 페이지로 자동 이동하지 않는 경우, 검색 결과 목록에서 해당 가게 클릭
        search_iframe()
        try:
            store_link = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, "span.place_bluelink.TYaxT")))
            driver.execute_script("arguments[0].scrollIntoView();", store_link)
            store_link.click()
            print(f"Clicked on store link for {store_name}")
            # 클릭 후 세부 정보 프레임으로 전환
            WebDriverWait(driver, 10).until(EC.frame_to_be_available_and_switch_to_it((By.ID, 'entryIframe')))
            entry_iframe()
            extract_details(store_name)  # 세부 정보 페이지에서 데이터 추출
        except Exception as e:
            print(f"Error clicking on store link: {e}")
            return  # 클릭 실패 시 함수 종료

# 모든 가게에 대해 크롤링 실행
for index, row in df_stores.iterrows():
    crawling_main(row['Store Name'], row['City'])

# 결과를 Excel 파일로 저장
naver_res.to_excel('./naver_crawling_result.xlsx', index=False)

# 웹드라이버 종료
driver.quit()
```
