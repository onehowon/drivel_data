```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

# ChromeDriver 설정
service = Service(ChromeDriverManager().install())
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=service, options=options)

# 시 정보
cities = [
    '목포시', '수원시', '성남시', '의정부시', '안양시', '부천시', '광명시', '평택시', '동두천시', '안산시', 
    '고양시', '과천시', '구리시', '남양주시', '오산시', '시흥시', '군포시', '의왕시', '하남시', '용인시', 
    '파주시', '이천시', '안성시', '김포시', '화성시', '광주시', '양주시', '포천시', '여주시', '청주시', 
    '충주시', '제천시', '천안시', '공주시', '보령시', '아산시', '서산시', '논산시', '계룡시', '당진시', 
    '전주시', '군산시', '익산시', '정읍시', '남원시', '김제시', '제주시', '서귀포시'
]

# 데이터를 저장할 리스트
all_store_names = []

for city in cities:
    try:
        driver.get('https://map.naver.com/')

        # 검색창 대기 및 입력
        search = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".input_search"))
        )
        search.send_keys(f"{city} 맛집")
        search.send_keys(Keys.ENTER)

        # 프레임 전환
        WebDriverWait(driver, 10).until(
            EC.frame_to_be_available_and_switch_to_it((By.ID, 'searchIframe'))
        )

        # 가게 이름을 저장할 리스트
        store_names = []

        def get_store_names():
            # 원하는 요소(가게 이름) 찾기
            titles = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".place_bluelink.TYaxT"))
            )

            for title in titles:
                store_names.append(title.text)

        # 첫 페이지 가게 이름 수집
        print(f"City: {city}, Page 1:")
        get_store_names()

        # 다음 페이지로 넘어가기
        num_pages = 5  # 예시로 5페이지까지만 검색
        
        for page in range(2, num_pages + 1):
            try:
                # 페이지 번호 버튼 클릭
                next_page_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, f"//a[@class='mBN2s ' and text()='{page}']"))
                )
                next_page_button.click()
                
                # 페이지가 완전히 로드될 때까지 대기
                time.sleep(2)  # 페이지 전환 대기
                
                # 프레임 다시 전환
                driver.switch_to.default_content()
                WebDriverWait(driver, 10).until(
                    EC.frame_to_be_available_and_switch_to_it((By.ID, 'searchIframe'))
                )

                # 현재 페이지 가게 이름 수집
                print(f"City: {city}, Page {page}:")
                get_store_names()
                
            except Exception as e:
                print(f"Failed to go to page {page} for {city}: {e}")
                break

        # 해당 위치의 모든 가게 이름을 전체 리스트에 추가
        all_store_names.extend(store_names)

        # 각 도시 별로 데이터를 DataFrame으로 변환
        df = pd.DataFrame(store_names, columns=["Store Name"])

        # CSV 파일로 저장 (각 도시 별로 저장)
        file_name = f"{city}_맛집.csv"
        df.to_csv(file_name, index=False, encoding='utf-8-sig')

        print(f"Data for '{city}' has been saved to '{file_name}'")

    except Exception as e:
        print(f"Error occurred while processing {city}: {e}")

# 드라이버 종료
driver.quit()

print("All data has been saved successfully")
```
